```python
import tkinter as tk
from tkinter import scrolledtext
import datetime
import json
import os

# File to store chat history
HISTORY_FILE = "chat_history.json"

# Simple memory
user_name = ""

# Save chat history
def save_chat(user, bot):
    data = {"user": user, "bot": bot}
    
    if os.path.exists(HISTORY_FILE):
        with open(HISTORY_FILE, "r") as file:
            chats = json.load(file)
    else:
        chats = []

    chats.append(data)

    with open(HISTORY_FILE, "w") as file:
        json.dump(chats, file, indent=4)

# Chatbot logic
def chatbot_response(user_input):
    global user_name
    user_input = user_input.lower()

    if "my name is" in user_input:
        user_name = user_input.split("my name is")[-1].strip().title()
        return f"Nice to meet you, {user_name}!"

    if "your name" in user_input:
        return "I am an Advanced Python AI Chatbot ðŸ¤–"

    if "time" in user_input:
        return "Current time is " + datetime.datetime.now().strftime("%H:%M:%S")

    if "date" in user_input:
        return "Today's date is " + datetime.datetime.now().strftime("%d-%m-%Y")

    if "hello" in user_input:
        if user_name:
            return f"Hello {user_name}! ðŸ˜Š"
        return "Hello! How can I help you?"

    # Simple math solving
    try:
        result = eval(user_input)
        return f"The answer is {result}"
    except:
        pass

    if "bye" in user_input:
        return "Goodbye! Have a great day! ðŸ‘‹"

    return "Sorry, I don't understand that yet."

# Send message
def send_message():
    user_input = entry.get()
    if user_input == "":
        return

    chat_area.insert(tk.END, "You: " + user_input + "\n")

    bot_reply = chatbot_response(user_input)
    chat_area.insert(tk.END, "Bot: " + bot_reply + "\n\n")

    save_chat(user_input, bot_reply)

    entry.delete(0, tk.END)

# Clear chat
def clear_chat():
    chat_area.delete("1.0", tk.END)

# GUI
window = tk.Tk()
window.title("Advanced AI Chatbot")
window.geometry("500x600")

chat_area = scrolledtext.ScrolledText(window, wrap=tk.WORD, font=("Arial", 12))
chat_area.pack(padx=10, pady=10, fill=tk.BOTH, expand=True)

entry = tk.Entry(window, font=("Arial", 12))
entry.pack(padx=10, pady=5, fill=tk.X)

send_button = tk.Button(window, text="Send", command=send_message)
send_button.pack(pady=5)

clear_button = tk.Button(window, text="Clear Chat", command=clear_chat)
clear_button.pack()

window.mainloop()
```
